```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('/content/hiring.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
import numpy as np

# Dictionary to convert words to numbers
word_to_num = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10
}

# Function to safely convert experience values
def convert_experience(val):
    if isinstance(val, str):
        val = val.lower().strip()
        return word_to_num.get(val, np.nan)  # if not found, return NaN
    return val  # if it's already a number or NaN

# Apply the conversion
df['experience'] = df['experience'].apply(convert_experience)

# Now fill missing values with median
df['experience'] = df['experience'].fillna(df['experience'].median())
```

```python
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
df['interview_score(out of 10)'] = df['interview_score(out of 10)'].fillna(df['interview_score(out of 10)'].mean())
```

```python
df.head()
df.isnull().sum()
```

```python
df
```

```python
# Independent variables (features)
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]

# Dependent variable (target)
y = df['salary($)']
```

```python
from sklearn.linear_model import LinearRegression

# Create model
model = LinearRegression()

# Train the model
model.fit(X, y)
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
prediction = model.predict([[2, 9, 6]])
print(f"Predicted Salary: ${prediction[0]:,.2f}")
```

```python
            # Predict salaries for entire dataset
df['Predicted Salary'] = model.predict(X)

# Save the result to a new CSV file
df.to_csv('salary_predictions.csv', index=False)
```
